```python
##Unemployment in India

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 2: Load and clean the dataset
df = pd.read_csv('Unemployment in India.csv')  # Make sure the filename is correct
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert date format
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
df = df.sort_values('date')

# Rename columns for easier access
df.rename(columns={
    'estimated_unemployment_rate_(%)': 'unemployment_rate',
    'estimated_labour_participation_rate_(%)': 'labour_participation_rate'
}, inplace=True)

# Step 3: Display stats
print("First 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())
print("\nMissing Values:\n", df.isnull().sum())

# Step 4: Plot overall unemployment trends (all regions)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='unemployment_rate', hue='region', legend=False)
plt.title('Unemployment Rate Over Time (All Regions)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Highlight COVID-19 Impact (Using region: India)
if 'India' in df['region'].unique():
    df_india = df[df['region'] == 'India'].copy()

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df_india, x='date', y='unemployment_rate', label='India')
    plt.axvspan(pd.to_datetime('2020-03'), pd.to_datetime('2021-12'), alpha=0.3, color='red', label='COVID-19 Period')
    plt.title('Impact of COVID-19 on Unemployment in India')
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("❗ Region 'India' not found in dataset. Skipping COVID-19 impact plot.")

# Step 6: Seasonal Decomposition (on 'India' if data is sufficient)
if 'India' in df['region'].unique():
    df_india = df[df['region'] == 'India'].copy()
    df_india.set_index('date', inplace=True)
    df_india = df_india.resample('M').mean(numeric_only=True)

    if len(df_india) >= 24:
        decomp = seasonal_decompose(df_india['unemployment_rate'], model='additive', period=12)
        decomp.plot()
        plt.suptitle('Seasonal Decomposition - India', fontsize=14)
        plt.tight_layout()
        plt.show()
    else:
        print("❗ Not enough data points for seasonal decomposition (need at least 24 months).")
else:
    print("❗ Region 'India' not found for decomposition.")

# Step 7: Policy Insight - Compare Pre/During/Post COVID-19
pre_covid = df[df['date'] < '2020-03-01']['unemployment_rate'].mean()
during_covid = df[df['date'].between('2020-03-01', '2021-12-31')]['unemployment_rate'].mean()
post_covid = df[df['date'] > '2021-12-31']['unemployment_rate'].mean()

print("\n📊 Average Unemployment Rate:")
print(f"Before COVID-19: {pre_covid:.2f}%")
print(f"During COVID-19: {during_covid:.2f}%")
print(f"After COVID-19: {post_covid:.2f}%")
```
